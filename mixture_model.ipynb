```python
elbo = pd.DataFrame({}, columns=range(1, 16), index=range(10))
```

```python
for cluster_num in elbo.columns:
    for epoch in elbo.index:
        d = data4stan.data4GaussianMixture(bayesian_path, cluster_num)
        fit = sm.vb(d, sample_file='dummy.csv', diagnostic_file=f'results/ppca/diagnostic_{cluster_num}.csv')
        elbo_val = data4stan.elbo_from_diagnostic(f'results/ppca/diagnostic_{cluster_num}.csv')
        elbo.loc[epoch, cluster_num] = elbo_val
```
